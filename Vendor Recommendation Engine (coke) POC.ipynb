```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
Data_Full = pd.read_csv('C:/Users/nikhi/Desktop/CocaCola/Agency_Data_Full.csv')
```

```python
#Creating Category filter

categories_dict = dict()


print("Select a category from the following by index")
for index, category in enumerate(list(map(lambda val: val[4:],set(Data_Full['Criteria'])))):
    categories_dict[index + 1] = category
    print(f'{index + 1} - {category}')
user_input = int(input())
#print(categories_dict[user_input])
first_filter = Data_Full[Data_Full['Criteria'].str.contains(categories_dict[user_input])]
first_filter


```

```python
#Creating Country/Region Filter

country_dict = dict()
print("Select a country from the following by index")
for index, country in enumerate(list(map(lambda val: val,set(first_filter['Geography Name'])))):
    country_dict[index + 1] = country
    print(f'{index + 1} - {country}')
user_input = int(input())
#print(categories_dict[user_input])
country_filter = first_filter[first_filter['Geography Name'].str.contains(country_dict[user_input])]
country_filter
```

```python
#Creating Brand filter

brand_dict = dict()
print("Select a brand from the following by index")
for index, Brand in enumerate(list(map(lambda val: val,set(country_filter['Brand'])))):
    brand_dict[index + 1] = Brand
    print(f'{index + 1} - {Brand}')
user_input = int(input())
#print(categories_dict[user_input])
brand_filter = country_filter[country_filter['Brand'].str.contains(brand_dict[user_input],na=False)]
brand_filter
```

```python
#sorting in descending values according to Asses criteria Score
brand_filter = brand_filter.sort_values(by='Assess Criteria Score', ascending=False)
brand_filter
```

```python
brand_filter.to_csv('C:/Users/nikhi/Desktop/CocaCola/VRE_output.csv', sep =',')
```
